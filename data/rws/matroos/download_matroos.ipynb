```python
import yarl
import pathlib

stations = [
    "vlissingen",
    "hoekvanholland",
    "ijmuiden",
    "denhelder",
    "harlingen",
    "delfzijl"
]

url = yarl.URL('http://noos.matroos.rws.nl/direct/get_series.php')


station = stations[0]

defaults = {
    "tstart": 201701010000,
    "tstop": 201901010100,
    "format": "netcdf"
}
params_list = [
    {
        "source": "observed",
        "unit": "waterlevel",
        "name": "waterlevel_observed"
    },
    {
        "source": "RWS_prediction",
        "unit": "waterlevel_surge",
        "name": "surge_forecast"
    },
    {
        "source": "RWS_prediction",
        "unit": "waterlevel",
        "name": "waterlevel_forecast"
    }
]

```

```python
for station in stations:
    for params in params_list:
        download_url = url.update_query(defaults).update_query(params).update_query({"loc": station})
        filepath = "{}_{}.nc".format(station, params["name"])
        !wget -c -O "$filepath" "$download_url"
```
