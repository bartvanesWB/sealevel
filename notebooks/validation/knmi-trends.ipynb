```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('~/src/sealevel/data/knmi/KNMI14_sealevel_scenarios_by_1year_3polfit.csv')
df = df.set_index('year')
df.tail()
```

```python
fig, ax = plt.subplots()
(100 * (df['low_05'] + df['low_95']) / 2).diff().plot(ax=ax, label='low mean rate', color='blue')
(100 * (df['high_05'] + df['high_95']) / 2).diff().plot(ax=ax,  label='high mean rate', color='green')

ax.fill_between(
    df['low_05'].index,
    df['low_05'].diff() * 100,
    df['low_95'].diff() * 100,
    label='low 5% rate',
    alpha=0.2,
    color='blue'
)

ax.fill_between(
    df['high_05'].index,
    df['high_05'].diff() * 100,
    df['high_95'].diff() * 100,
    label='high 5% rate',
    alpha=0.2,
    color='green'
)
ax.set_ylabel('Sea-level rise rate [cm/century]')
ax.legend()
```
