```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names = ['year', 'land', 'land_smooth', 'ocean', 'ocean_smooth']
ocean_land_df = pd.read_csv('/Users/baart_f/src/sealevel/data/nasa/ocean_land_temperature.csv', skiprows=2, names=names)
```

```python
fig, ax = plt.subplots(figsize=(13, 8))
ax.plot(ocean_land_df.year, ocean_land_df.ocean, label='ocean')
ax.plot(ocean_land_df.year, ocean_land_df.land, label='land')
ax.set_xlabel('time')
ax.set_ylabel('temperature anomaly [degrees Celcius]')
ax.legend(loc='best')
```

```python
fig, ax = plt.subplots(figsize=(13, 8))
ax.plot(np.diff(ocean_land_df.land), np.diff(ocean_land_df.ocean), 'k.')
# explained variance of ocean temperature by land temperature
np.corrcoef(
    np.diff(ocean_land_df.land), 
    np.diff(ocean_land_df.ocean)
)**2
```
